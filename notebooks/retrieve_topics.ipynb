```python
import pandas as pd
```

```python
import os
import sys

current_dir = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
parent_dir = os.path.abspath(os.path.join(current_dir, ".."))
sys.path.insert(0, parent_dir)

from src.topics_utilities import find_wikidata_entity, get_subfield, get_subfield_by_author
```

```python
df_papers = pd.read_csv("../data/papers/papers.csv")
df_papers
```

```python
df_bicocca = pd.read_csv("../data/authors/authors_internal_prova.csv")
df_bicocca.head()
```

```python
topics = []

for index, row in df_papers.iterrows():
    doi = row["doi"]
    subfield = get_subfield(doi)
    if subfield not in topics and subfield:
        topics.append(subfield)

print(topics)
```

```python
for index, row in df_bicocca.iterrows():
    orcid = row["orcid"]
    subfields = get_subfield_by_author(orcid)
    if subfield not in topics and subfield:
        for subfield in subfields:
            topics.append(subfield)

print(topics)
```

```python
df_topics = pd.DataFrame(topics, columns=["topic", "id"])
df_topics = df_topics.dropna()
df_topics
```

```python
df_topics["api_url"] = df_topics["id"].apply(
    lambda x: x.replace("https://openalex.org", "https://api.openalex.org") if pd.notna(x) else None
)
df_topics
```

```python
df_topics["wikidata_uri"] = df_topics["api_url"].apply(find_wikidata_entity)
df_topics
```

```python
df_topics[["topic", "wikidata_uri"]].to_csv("../data/topics/topics.csv", index=False)
```
