```python
import pandas as pd
import requests
import time
import ast
```

```python
import os
import sys

current_dir = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
parent_dir = os.path.abspath(os.path.join(current_dir, ".."))
sys.path.insert(0, parent_dir)

from src.papers_utilities import get_works
from src.topics_utilities import get_subfield
```

```python
df_authors = pd.read_csv("../data/authors/authors_complete_ds.csv")
```

```python
df_papers = pd.DataFrame(columns=["doi", "year", "authors", "author_orcids", "type", "topics"])
```

```python
for index, row in df_authors.iterrows():
    family_name = row['Last name']
    given_name = row['Name']
    orcid = row['orcid']
    papers = get_works(orcid, given_name, family_name)
    df_papers = pd.concat([df_papers, papers], ignore_index=True)
    time.sleep(5)

df_papers
```

```python
# Removing duplicates
temp = df_papers[~df_papers.duplicated(subset=["doi"], keep="first")]
df_papers_no_duplicate = temp[~temp.duplicated(subset=["title"], keep="first")]

df_papers_no_duplicate
```

```python
for index, row in df_papers_no_duplicate.iterrows():
    authors = ast.literal_eval(row["author_orcids"])
    if len(authors) > 10:
        df_papers_no_duplicate = df_papers_no_duplicate.drop(index)

df_papers_no_duplicate
```

```python
df_papers_no_duplicate.to_csv("../data/papers/papers.csv", index=False)
```
